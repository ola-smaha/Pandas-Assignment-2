So, we might conclude that the year in which the house is built does not make a big difference in price based on the variables used in this analysis.**

Distance Function:

````python
from math import sin, cos, acos, radians
def dist(lat1,lon1,lat2, lon2):
    lat11 = radians(lat1)
    lat22 = radians(lat2)
    lon11 = radians(lon1)
    lon22 = radians(lon2)
    return acos(sin(lat11)*sin(lat22)+cos(lat11)*cos(lat22)*cos(lon22-lon11))*6371
#testing
#print(dist(47.7210,-122.319,47.7379,-122.233))
#Output: 6.7011090149458985
````

## 1. Distance by id

````python
id_list = list(df['id'])

def dist_id(id1,id2):
    coordinates1 = list(df.loc[(df.id== id1),['lat','long']].values)
    coordinates2 = list(df.loc[(df.id== id2),['lat','long']].values)
    distance = dist(coordinates1[0][0],coordinates1[0][1],coordinates2[0][0],coordinates2[0][1])
    return distance
    
id1 = int(input('Enter id1: '))
while id1 not in id_list:
    id1 = int(input('Enter valid id: '))
id2 = int(input('Enter id2: '))
while id2 not in id_list:
    id2 = int(input('Enter valid id: '))

print(dist_id(id1,id2))
````

Looking at id's of most expensive houses:

````python
df2 = df[['id','price']].sort_values('price', ascending = False).copy()
df2.head(10)
````

Distance between most expensive house and second most expensive house:

````python
print(dist_id(df2.iloc[0,0],df2.iloc[1,0]))
````

Distance between most expensive house and third most expensive house:

````python
print(dist_id(df2.iloc[0,0],df2.iloc[2,0]))
````

Looking at id's of cheapest houses:

````python
df3 = df[['id','price']].sort_values('price').copy()
df3.head(10)
````

Distance between most expensive house and cheapest house:

````python
print(dist_id(df2.iloc[0,0],df3.iloc[0,0]))
````

Distance between most expensive house and third cheapest house:

````python
print(dist_id(df2.iloc[0,0],df3.iloc[2,0]))
````

Distance between cheapest house and second cheapest house:

````python
print(dist_id(df3.iloc[0,0],df3.iloc[1,0]))
````

Distance between cheapest house and third cheapest house:

````python
print(dist_id(df3.iloc[0,0],df3.iloc[2,0]))
````

Distance between most cheapest house and fourth cheapest house:

````python
print(dist_id(df3.iloc[0,0],df3.iloc[3,0]))
````

### Insights:

```It seems like the most expensive houses are situated close to each other and the cheapest houses are much farther away from them. Moreover and interestingly enough, the cheaper houses are also far from each other, at least farther from each other than the expensive houses are from each other.```

## .iat

```
        DataFrame.iat
```
**Access a single value for a row/column pair by integer position.<br>
Similar to iloc, in that both provide integer-based lookups. Use iat if you only need to get or set a single value in a DataFrame or Series.**

iloc also works for that purpose, but iat is faster, while iloc can handle more complex scenarios.

````python
df2.iat[0,0]
````

````python
df2.iloc[0,0]
````

The following code raises an error, that's because the integer position is out of bound.
```
        df3.iat[4,7]
```

````python
# Example on original dataframe
df.iat[2352,16]
````

## .at

```
        DataFrame.at
```
**Access a single value for a row/column label pair.<br>
Similar to loc, in that both provide label-based lookups. Use **at** if you only need to get or set a single value in a DataFrame or Series.**

````python
df.at[4, 'id']
````
